```python
d = {"1": "X", "2": "X", "3": "X", "4": "X", "5": "X", "6": "X", "7": "X", "8": "X", "9": "X"}

print(d)
```

```python
def board():

    print(d["1"], "|" ,d["2"], "|" ,d["3"])
    print("---------")
    print(d["4"], "|" ,d["5"], "|" ,d["6"])
    print("---------")
    print(d["7"], "|" ,d["8"], "|" ,d["9"])
```

```python
board()
```

```python
def clear_board():

    x = -1

    while x < 9:
        x += 1

    
        d[str(x)] = " "
```

```python
def check_winner():

    if d["1"] ==  d["2"]  and   d["1"] ==  d["3"]  and    d["2"] ==  d["3"]:
        if d["1"] == "X" and d["2"] == "X" and d["3"] == "X":
            print("player(X) won")
        elif d["1"] == "O" and d["2"] == "O" and d["3"] == "O":
            print("player(O) won")
            
    

    elif d["4"] ==  d["5"]  and   d["4"] ==  d["6"]  and    d["5"] ==  d["6"]:
        if d["4"] == "X" and d["5"] == "X" and d["6"] == "X":
            print("player(X) won")    
        elif d["4"] == "O" and d["5"] == "O" and d["6"] == "O":
            print("player(O) won")
            
    

    elif d["7"] ==  d["8"]  and   d["7"] ==  d["9"]  and    d["8"] ==  d["9"]:
        if d["7"] == "X" and d["8"] == "X" and d["9"] == "X":
            print("player(X) won")    
        elif d["7"] == "O" and d["8"] == "O" and d["9"] == "O":
            print("player(O) won") 
            
    

    elif d["1"] ==  d["4"]  and   d["1"] ==  d["7"]  and    d["4"] ==  d["7"]:
        if d["1"] == "X" and d["4"] == "X" and d["7"] == "X":
            print("player(X) won")   
        elif d["1"] == "O" and d["4"] == "O" and d["7"] == "O":
            print("player(O) won")
            
    

    elif d["2"] ==  d["5"]  and   d["2"] ==  d["8"]  and    d["5"] ==  d["8"]:
        if d["2"] == "X" and d["5"] == "X" and d["8"] == "X":
            print("player(X) won")   
        elif d["2"] == "O" and d["5"] == "O" and d["8"] == "O":
            print("player(O) won")
    

    elif d["3"] ==  d["6"]  and   d["3"] ==  d["9"]  and    d["6"] ==  d["9"]:
        if d["3"] == "X" and d["6"] == "X" and d["9"] == "X":
            print("player(X) won")  
        elif d["3"] == "O" and d["6"] == "O" and d["9"] == "O":
            print("player(O) won") 
            
    

    elif d["1"] ==  d["5"]  and   d["1"] ==  d["9"]  and    d["5"] ==  d["9"]:
        if d["1"] == "X" and d["5"] == "X" and d["9"] == "X":
            print("player(X) won")   
        elif d["1"] == "O" and d["5"] == "O" and d["9"] == "O":
            print("player(O) won") 
            
    

    elif d["3"] ==  d["5"]  and   d["3"] ==  d["7"]  and    d["5"] ==  d["7"]:
        if d["3"] == "X" and d["5"] == "X" and d["7"] == "X":
            print("player(X) won")
        elif d["3"] == "O" and d["5"] == "O" and d["9"] == "7":
            print("player(O) won")
            
    else:
        print("it is a tie: try again!!!")
            
    
```
